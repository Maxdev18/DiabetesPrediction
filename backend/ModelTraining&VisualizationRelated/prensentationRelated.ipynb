```python
import bars
import pandas as pd

file_path = 'diabetes_012_health_indicators_BRFSS2015.csv'
data = pd.read_csv(file_path)

diabetes_counts = data['Diabetes_012'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
bar_container = plt.bar(diabetes_counts.index, diabetes_counts.values, color=['blue','red', 'green'])
plt.title('Distribution of Diabetes Diagnosis')
plt.xlabel('Diabetes Status')
plt.ylabel('Count')
plt.xticks(ticks=[0,1, 2], labels=['No Diabetes (0)', 'Pre-Diabetes(1)', 'Diabetes (2)'], rotation=0)

# Add the text annotations on the bars
for bar in bar_container:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/3, yval, round(yval, 3), va='bottom', ha='center')


plt.savefig('diabetes_distribution_chart.png', bbox_inches='tight')
plt.show()
```
